```python
#append tables
list_of_dfs = [contrib1_df,contrib2_df,contrib3_df,contrib4_df,contrib5_df,contrib6_df,contrib7_df,contrib8_df,
              contrib9_df,contrib10_df,contrib11_df,contrib12_df,contrib13_df]
combined_contribs = pd.concat(list_of_dfs, ignore_index=True)
combined_contribs.tail(100)
```

```python
#show columns and data types
combined_contribs.dtypes
```

```python
#drop all non-candidate committee types, ie PACS, and also judicial candidates
select_contribs = combined_contribs.loc[combined_contribs["filerTypeCd"] =="COH"]
select_contribs.head(100)

```

```python
#only keep necessary columns
clean_contribs = select_contribs[['receivedDt', 'filerIdent', 'filerName', 'contributionDt', 'contributionAmount', 'contributorPersentTypeCd', 
                                'contributorNameOrganization', 'contributorNameLast','contributorNameFirst','contributorStreetCity','contributorStreetStateCd', 'contributorStreetPostalCode',
                                'contributorEmployer']]
clean_contribs
```

```python
#rename columns
renamed_df = clean_contribs.rename(columns={"receivedDt":"rpt_received_date", "filerIdent":"filer_ID", "filerName":"filer_name",
                                           "contributionDt": "contrib_date", "contributionAmount":"contrib_amount",
                                           "contributorPersentTypeCd": "contributor_type", "contributorNameOrganization": "contributor_org",
                                            "contributorNameLast":"contributor_last_name",
                                           "contributorNameFirst":"contributor_first_name", "contributorStreetCity": "contributor_city",
                                           "contributorStreetStateCd":"contributor_state", "contributorStreetPostalCode":"contributor_zipcode",
                                           "contributorEmployer":"contributor_employer"})
renamed_df.head()
```

```python
#separate the df based on whether the contributor is an entity or individual
indiv_contribs = renamed_df.loc[renamed_df["contributor_type"] =="INDIVIDUAL"]
org_contribs = renamed_df.loc[renamed_df["contributor_type"] =="ENTITY"]
org_contribs.head()
```

```python
#drop even more columns, for ease of use in sql
indiv_contribs = indiv_contribs[['filer_name', 'contributor_last_name', 'contributor_first_name', 'contrib_amount',
                                 'contrib_date', 'filer_name']]
```

```python
#drop even more columns, for ease of use in sql
org_contribs = org_contribs[['contributor_org', 'contrib_amount', 'contrib_date', 'filer_name']]
```

```python
#group by contributors
grouped_indiv_contribs = indiv_contribs.groupby(['contributor_last_name', 'contributor_first_name'])
grouped_indiv_contribs.head()

grouped_org_contribs = org_contribs.groupby(['contributor_org'])
grouped_org_contribs.head()
```

```python
#import 2 contracting csv files & read in
load_box123 = "../project/1295 Certificates/Box123.csv"
read_box123 = pd.read_csv(load_box123)
load_parties = "../project/1295 Certificates/Interested Party.csv"
read_parties = pd.read_csv(load_parties)
```

```python
#drop non-state level entities in Box123 table
select_box123 = read_box123.loc[read_box123["governmentTypeCd"] =="GOVT_ENTITY"]
select_box123.tail()
```

```python
#rename columns in Box 123 dataframe
renamed_box123 = select_box123.rename(columns={"acknowledgeDate":"ReceivedDate","businessCityName":"bizCity",
                                              "businessStateCode":"bizState", "goodsServicesDescr": "description"})
renamed_box123.head()
```

```python
#drop unnessary columns in box123
clean_box123 = renamed_box123[['certificateNumber','reportNumber','ReceivedDate','governmentName',
                               'businessName', 'bizCity','bizState','description']]
clean_box123.head()
```

```python
#removing certain entities: cities, counties & school districts. They shouldn't be type=govt_entity, but some are
#also couldn't get them to string together, so did them separately, the dumb way


df1 = clean_box123[~clean_box123["governmentName"].str.contains("University", na=False)]
df2 = df1[~df1["governmentName"].str.contains("College", na=False)]
df3 = df2[~df2["governmentName"].str.contains("County", na=False)]
df4 = df3[~df3["governmentName"].str.contains("City", na=False)]
df5 = df4[~df4["governmentName"].str.contains("ISD", na=False)]
df6 = df5[~df5["governmentName"].str.contains("District", na=False)]


df6.head(5000)
```

```python
read_parties.head()
```

```python
#keep only necessary columns in parties df
clean_parties = read_parties[['certificateNumber','reportNumber','partyTypeCode','partyOrgName','partyLastName',
                              'partyFirstName', 'partyCity', 'partyStateCode']]
clean_parties.head()
```

```python
#rename columns in parties df
renamed_parties = clean_parties.rename(columns={"partyOrgName":"businessName",
                                              "partyCity":"OrgCity", "partyStateCode": "OrgState"})
renamed_parties.head()
```

```python
#merge parties and box123 
merged_contracts = pd.merge(renamed_parties, df6, on="businessName", how = "left")
merged_contracts
```

```python
grouped_orgs = merged_contracts.groupby(['businessName'])
grouped_orgs.head()
```

```python
#drop a bunch of columns, for ease of use in sql
merged_contracts = merged_contracts[['partyTypeCode','businessName','partyLastName','partyFirstName',
                                     'certificateNumber_x','reportNumber_x']]
```

```python
grouped_indivs = merged_contracts.groupby(['partyLastName', 'partyFirstName'])
grouped_indivs.head()
```
